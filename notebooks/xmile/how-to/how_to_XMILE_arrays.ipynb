```python
from BPTK_Py import bptk
bptk = bptk()
```

```python
bptk.plot_scenarios(
    scenario_managers=["smArrayedVariables"],
    scenarios=["base"], 
    equations=["totalInventory"]
)
```

```python
bptk.plot_scenarios(
    scenario_managers=["smArrayedVariables"],
    scenarios=["base"], 
    equations=["smallestGermanInventory","largestGermanInventory"]
)
```

```python
bptk.plot_scenarios(
    scenario_managers=["smArrayedVariables"],
    scenarios=["base"], 
    equations=["averageInventory","averageInventoryUsingSize"]
)
```

```python
bptk.plot_scenarios(
    scenario_managers=["smArrayedVariables"],
    scenarios=["base"], 
    equations=["inventory[1,germany]","inventory[1,england]"]
)
```
